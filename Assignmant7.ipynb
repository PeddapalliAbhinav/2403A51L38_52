```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset (World Population across decades)
data = {
    "Year": [1960, 1970, 1980, 1990, 2000, 2010, 2020],
    "Population": [3.0, 3.7, 4.4, 5.3, 6.1, 6.9, 7.8]  # in billions
}

# Convert to DataFrame (simulate CSV load)
df = pd.DataFrame(data)

# Plot line chart
plt.figure(figsize=(8,5))
plt.plot(df["Year"], df["Population"], marker="o", linestyle="-", color="b", label="Population Growth")

# Labels and title
plt.xlabel("Year")
plt.ylabel("World Population (Billions)")
plt.title("World Population Growth Over Decades")
plt.legend()
plt.grid(True)

# Show chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create sample dataset
data = {
    "Date": pd.date_range(start="2023-01-01", periods=10, freq="M"),
    "Close": [130, 135, 138, 140, 145, 150, 155, 160, 158, 162]
}

df = pd.DataFrame(data)

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(df["Date"], df["Close"], marker="o", color="blue", linewidth=2, label="AAPL Close Price")

plt.xlabel("Date")
plt.ylabel("Close Price (USD)")
plt.title("Apple Stock Price Trend (Sample Data)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

plt.show()
```
